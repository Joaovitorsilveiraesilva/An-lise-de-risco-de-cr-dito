Como queremos testar apenas a multicolinearidade, vamos excluir FONDKAPREMONT HOUSETYPE TOTALAREA WALLSMATERIA EMERGENCYSTATE por serem variáveis isoladas.

```python
index_mult = train_mam.str.split('_').str[0].value_counts()[:13].index.values
np.in1d(train.columns.str.split('_').str[0] , index_mult)
train.columns.values[np.in1d(train.columns.str.split('_').str[0] , index_mult)]
train_mam = train[train.columns.values[np.in1d(train.columns.str.split('_').str[0] , index_mult)]]
```

```python
def var_mam(x):
    return train[train_mam.columns[train_mam.columns.str.startswith(var_index[x])]]
```

```python
var_mam(0).corr()
```

```python
np.average(var_mam(0).corr().values)
```

```python
def corr90(x):
    return x>0.9
```

```python
var_mam(0).corr().apply(corr90).values
```

```python
np.all(var_mam(0).corr().apply(corr90).values)
```

```python
var_corr90 = []
avg_corr = []
corr_var = pd.DataFrame()
for i, k in enumerate(var_index.unique()):
    if np.all(var_mam(i).corr().apply(corr90).values):
        var_corr90.append(k)
        avg_corr.append(np.average(var_mam(i).corr().values))
corr_var['var'] = var_corr90
corr_var['media_corr'] = avg_corr
        
    
```

```python
corr_var
```

As variáveis acima possuem uma correlação muito próxima de 1 em média entre suas modas, médias e mediana.

Critério de escolha de variável, podemos pegar somente a variável que tem maior correlação em módulo com TARGET

```python
corr_var['var']
```

```python
def var_mam_excl(x):
    return train[train_mam.columns[train_mam.columns.str.startswith(corr_var['var'][x])]].join(train['TARGET'])
```

```python
np.absolute(var_mam_excl(0).corr()['TARGET']).sort_values(ascending=False)
```

```python
var_para_excluir = []
for i, k in enumerate(corr_var['var']):
    var_para_excluir.extend(np.absolute(var_mam_excl(i).corr()['TARGET']).sort_values(ascending=False)[-2:].index)
```

```python
var_para_excluir
```

As variáveis acima possuem alta correlação entre suas médias, medianas e modas. Então excluímos as variáveis com menor correlação com TARGET

```python
novas_colunas = train.columns[~(np.in1d(train.columns , var_para_excluir))]
```

```python
train = train[novas_colunas]
```
